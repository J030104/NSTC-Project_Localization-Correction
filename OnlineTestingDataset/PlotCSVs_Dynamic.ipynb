```python
import os
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
current_directory = os.getcwd()
directory = r'Dynamic_Dataset'
output_directory = r'Dynamic_Plots'
if not os.path.exists(output_directory):
    os.makedirs(output_directory)
```

```python
csv_files = [file for file in os.listdir(os.path.join(current_directory, directory)) if file.endswith('.csv')]
print(csv_files)
# csv_file = os.path.join(current_directory, directory, csv_files[1])
# csv_file = os.path.join(current_directory, directory, 'admin2ib1.txt.csv')
```

```python
def plotDynamic(csv_file, output_directory):
    x_values = []
    y_values = []

    # Check if the file is empty
    if os.path.getsize(csv_file) > 0:
        df = pd.read_csv(csv_file, header=None)  # Don't use header
        x_values.extend(df.iloc[:, 0].tolist())  # Extract first column as x
        y_values.extend(df.iloc[:, 1].tolist())  # Extract second column as y
    else:
        print(f"Warning: File {csv_file} is empty and will be skipped.")

    if len(x_values) > 0 and len(y_values) > 0:
        # Plotting the data
        plt.figure(figsize=(8, 6))
        plt.title('2D Plot of Data: ' + os.path.basename(csv_file))
        plt.xlabel('X Axis')
        plt.ylabel('Y Axis')
        plt.legend()
    
        plt.scatter(x_values, y_values, label='Data Points')

        output_path = os.path.join(current_directory, output_directory)
        output_file = os.path.join(output_path, f'{os.path.basename(csv_file)}.png')
        plt.savefig(output_file)
        # plt.show()
    else:
        print("Error: No valid data points found to plot.")
```

```python
for csv_file in csv_files:
    file = os.path.join(current_directory, directory, csv_file)
    plotDynamic(file, output_directory)
```
